# Preparation

## Map
1. works on **Series(columns)**. 
2. replace values according to the given dictionary.

choose `map` over apply for a single column transformation

```python
df['class'] = df['class'].map({'Iris-setosa':'SET', 'Iris-virginica':'VIR', 'Iris-versicolor':'VER'})
df.head()
```

## Apply
1. works with both DataFrames and Series.

```python
# Apply on Series
df['wide petal'] = df['petal width'].apply(lambda v: 1 if v >= 1.3 else 0)
df.head()
```

```python
# Apply on a DataFrames
df['petal area'] = df.apply(lambda r: r['petal length'] * r['petal width'], axis=1)
df.head()
```

Summary: 
1. Apply 函数无论是应用于Series 和dataframe的区别在于可引用的数据范围，df.apply当axis=1时，
    对可以获取到每一行每一列的数据. 而应用于Series时，只能获取到当前Series columns的某一行的值
2. Apply 函数无论是对于Series 还是dataframe，产生的结果都是一个Series，这个Series能够被引用去替换原始的dataframe的某一列.

## Applymap
1. perform a function across all data cells.
2. Common uses for applymap are to **transform each cell based on meeting some conditional criteria.**

```python
# Type checking (isinstance)
df_log = df.applymap(lambda v: np.log(v) if isinstance(v, float) else v)
df_log.head()
```

## Statsmodels

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(df['sepal width'][:50], df['sepal length'][:50])
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Sepal Width')
ax.set_title('Setosa Sepal Width vs. Sepal Length', fontsize=14, y=1.02)
```

```python
import statsmodels.api as sm

y = df['sepal length'][:50]
x = df['sepal width'][:50]
X = sm.add_constant(x)

results = sm.OLS(y, X).fit()
print(results.summary())
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(x, results.fittedvalues, label='regression line')
ax.scatter(x, y, label='data point', color='r')
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Sepal Width')
ax.set_title('Setosa Sepal Width vs. Sepal Length', fontsize=14, y=1.02)
ax.legend(loc=2)
```

## Scikit-earn

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
```

```python
clf = RandomForestClassifier(max_depth=5, n_estimators=10)

X = df.iloc[:, :4]
y = df.iloc[:, 4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
rf = pd.DataFrame(list(zip(y_pred, y_test)), columns=['predicted', 'actual'])
rf['correct'] = rf.apply(lambda r: 1 if r['predicted'] == r['actual'] else 0, axis=1)
rf.head()
```

```python
accuracy = rf['correct'].sum() / rf['correct'].count()
print(accuracy)
```

#### see which features gave us the most discriminative or predictive power:

1. The random forest has a method called .feature_importances_ that returns the **relative power of the feature to split at the leaves**.
2. This number will always **sum to 1**.

```python
f_importances = clf.feature_importances_
f_names = df.columns[:4]
f_std = np.std([tree.feature_importances_ for tree in clf.estimators_], axis=0)

# Sort zip
zz = zip(f_importances, f_names, f_std)
zzs = sorted(zz, key=lambda x: x[0], reverse=True)

imps = [x[0] for x in zzs]
labels = [x[1] for x in zzs]
errs = [x[2] for x in zzs]

plt.xticks(range(len(f_importances)), labels)
plt.bar(range(len(f_importances)), imps, color='r', yerr=errs, align='center')
```

## Suport vector machine (SVM)

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
```

```python
clf = OneVsRestClassifier(SVC(kernel='linear'))

X = df.iloc[:, :4]
y = np.array(df.iloc[:, 4])#.astype(str)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

svm = pd.DataFrame(list(zip(y_pred, y_test)), columns=['predicted', 'actual'])
svm['correct'] = svm.apply(lambda r: 1 if r['predicted'] == r['actual'] else 0, axis=1)
svm.head()
```

```python
accuracy = svm['correct'].sum() / svm['correct'].count()
print(accuracy)
```
